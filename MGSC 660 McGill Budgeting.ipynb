# Part A

```python
from pulp import *

#init and set lower bounds
x,y,z,g,a = LpVariable('x',0, cat="Binary"), LpVariable('y',0,cat="Binary"), LpVariable('z',0,cat="Binary"), LpVariable('g',0,cat="Binary"), LpVariable('a',0,cat="Binary")

prob = LpProblem("Capital_budgeting_at_the_McGill_Corporation", LpMaximize)

objective_fn = 10*x + 17*y + 16*z + 8*g + 14*a

prob += objective_fn

#constraints
prob += 48*x + 96*y + 80*z + 32*g + 64*a <= 160

prob.solve()
print(f"x: {value(x)}, y: {value(y)}, z: {value(z)}, g: {value(g)}, a: {value(a)}")
```

# Part B

```python
from pulp import *

#init and set lower bounds
x,y,z,g,a = LpVariable('x',0, cat="Binary"), LpVariable('y',0,cat="Binary"), LpVariable('z',0,cat="Binary"), LpVariable('g',0,cat="Binary"), LpVariable('a',0,cat="Binary")

prob = LpProblem("Capital_budgeting_at_the_McGill_Corporation", LpMaximize)

objective_fn = 10*x + 17*y + 16*z + 8*g + 14*a

prob += objective_fn

#constraints
prob += 48*x + 96*y + 80*z + 32*g + 64*a <= 160
prob += x + z + g >= 2
prob += y + a <= 1
prob += x + z >= a
prob += x + z + a <= 2
prob += g <= x
prob += g <= y
prob += x + y - 1 <= g

prob += x <= z + g + a
prob += y <= z + g + a

prob.solve()
print(f"x: {value(x)}, y: {value(y)}, z: {value(z)}, g: {value(g)}, a: {value(a)}")
```
